## Description of the problem and a discussion of the background

I have recently moved to Nashville, Tennessee and would like to open a restaurant. I would like to know which neighborhoods have the highest customer ratings, or have the largest population and most restaurant reviews. Does a certain neighborhood have more visitors than others, thus more likely to attract new customers?

## Description of how the data will be used to solve the problem

The data will be collected and analyzed to determine if a certain neighborhood has higher customer ratings than others, and which neighborhood has the highest reviews. If there is not a consensus as to which neighborhood has the highest ratings and the most reviews, the information will be analyzed to determine if a certain neighborhood can be determined as a "better" option. If not, choose the neighborhood with the higher ratings, or more visitors, depending on the style of restaurant to be opened. The style of restaurant will not be specified so that is as far as this analysis will take us.
